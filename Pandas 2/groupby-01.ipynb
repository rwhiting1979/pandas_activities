```python
# Import Libraries and Dependencies
# Import necessary libraries and dependencies
import pandas as pd
from pathlib import Path
%matplotlib inline
```

```python
# Create a Path to the File Using Pathlib
# Use the Pathlib libary to set the path to the CSV
csv_path = Path('cleaned_people_data.csv')
```

```python
# Use the file path to read the CSV into a DataFrame and display a few rows
people_df = pd.read_csv(csv_path)
people_df.head()
```

```python
# Group DataFrame by Occupation and perform count aggregation
# Group by `Occupation` and perform count
people_df.groupby('Occupation').count()
```

```python
# Group DataFrame by Occupation and Calculate Average Salary and Age
# Calculate average Salary and Age for each Occupation
people_df.groupby('Occupation').mean()
```

```python
# Group By Occupation and Gender Columns, then Calculate Average Salary and Age
# Group by Occupation and Gender columns
people_df.groupby(['Occupation','Gender']).mean()
```
